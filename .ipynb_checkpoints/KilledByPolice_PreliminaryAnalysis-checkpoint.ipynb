# KilledByPolice.net Preliminary Data Analysis
The data within this notebook was gathered from the tables at killedbypolice.net and has been cleaned and organized into a csv file. At this time, one of the attributes in the table is unknown. It is marked with an asterisk (*) on the KilledByPolice.net website.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
kbpn = pd.read_csv("Data/kbpn_data.csv")
```

```python
# code to count frequency of age ranges
one_to_twelve = 0
thirteen_to_nineteen = 0
twenties = 0
thirties = 0
forties = 0
fifties = 0
sixties = 0
seventies = 0
eighties = 0
nineties = 0

kbpn_age_list = []

for item in kbpn["Age"].dropna():
    if item.isdigit() == True:
        # convert to integer
        item = int(item)
        kbpn_age_list.append(item)
        if 0 <= item <= 12:
            one_to_twelve += 1
        if 13 <= item <= 19:
            thirteen_to_nineteen +=1
        if 20 <= item <= 29:
            twenties += 1
        if 30 <= item <= 39:
            thirties += 1
        if 40 <= item <= 49:
            forties += 1
        if 50 <= item <= 59:
            fifties += 1
        if 60 <= item <= 69:
            sixties += 1
        if 70 <= item <= 79:
            seventies += 1
        if 80 <= item <= 89:
            eighties += 1
        if 90 <= item <= 99:
            nineties += 1

kbpn_age_dict = {"1-12": one_to_twelve, "13-19": thirteen_to_nineteen, "20-29": twenties, "30-39": thirties, \
                  "40-49": forties, "50-59": fifties, "60-69": sixties, "70-79" : seventies, "80-89": eighties, \
                  "90-99": nineties}
kbpn_age_dict
```

```python
df_age_kbpn = pd.DataFrame(kbpn_age_list, columns=["Age"])
df_age_kbpn.describe()
```

```python
keys = kbpn_age_dict.keys()
values = kbpn_age_dict.values()

plt.bar(keys, values)
plt.xlabel("Age")
plt.ylabel('Frequency')
plt.title('Frequency of Police Fatalities by Age')
plt.xticks(rotation=30)
```

```python
labels = '0-12', '13-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79', '80-89', '90-99'
total = 5+287+1478+1596+995+654+254+66+18+1
sizes = [5/total, 287/total, 1478/total, 1596/total, 995/total, 654/total, 254/total, 66/total, 1/total]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(labels, loc='best', bbox_to_anchor=(-0.1, 1.),
           fontsize=12)
plt.title('Percentage of Police Fatalities by Age')
plt.show()
```

## Summary of Insights into the KilledByPolice.Net Data

Age: 5354/5596 entries reported age. The mean age is 37.111505. The standard deviation is 13.172301. The minimum is 3 according to the pandas description function, but a 9 month old was reported. However, "9 mo." is not an integer or float. The median age is 35. The maximum age is 91. 0.9% of the victims of police fatalities fall between the 1-12 age range (5 cases). 5.4% of the victims fall between the 13-19 age range (287 cases). 27.6% of the victims of police fatalities fall between the 20-29 age range (1478 cases). 29.8% of the victims fall between the 30-39 age range (1596 cases). 18.6% fall between 40-49 (995 cases). 12.2% fall between 50-59 (654 cases). 4.7% fall between 60-69 (254 cases). 1.2% fall between 70-79 (66 cases). 0.3% fall between 80-89 (18 cases). 0.01% fall between 90-99 (1 case).

Gender: 95.5% of the victims reported were male (5343 cases). 4.5% of the victims reported were female (251 cases).

Race: 50.6% of victims reported were white. 26.5% of victims were Black. 17.9% of victims were Hispanic. 1.9% of victims were Asian/Pacific Islander. 1.5% of victims were Native American. 0.9% of victims fall into the "other" racial category. It is unclear what L, I, and M stand for in this dataset.

State: The top twenty states in the dataset are: CA (825 cases), TX (498 cases), FL (373 cases), AZ (263 cases), CO (201 cases), GA (188 cases), OK (175 cases), OH (161 cases), NC (160 cases), WA (155 cases), MO (146 cases), TN (142 cases), PA (114 cases), LA (113 cases), AL (110 cases), NM (109 cases), IL (108 cases), NY (104 cases), NV (99 cases), IN (98 cases).
