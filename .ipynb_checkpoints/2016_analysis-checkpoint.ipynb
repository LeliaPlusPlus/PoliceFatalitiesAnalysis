# Preliminary Data Analysis for 2016 Data in The Washington Post, The Guardian, and Fatal Encounters Data Sets

Each data sets contains data for certain years: Fatal Encounters (2000-2020), The Washington Post (2015-2020), The Guardian (2016). Note that The Guardian has data for 2015, but it is not publicly accessible in a downloadable format (i.e. csv). Data for the common year 2016 is to be compared to understand how similar or dissimilar the data for 2016 is in each dataset.

Of course, this comparison faces challenges in terms of each dataset tracking differing attributes and differing categories in each attribute. However, there are some similar attributes and categories across datasets, for example, age and gender. 

#### Data Sources
Fatal Encounters: https://docs.google.com/spreadsheets/d/1dKmaV_JiWcG8XBoRgP8b4e9Eopkpgt7FL7nyspvzAsE/edit#gid=0
The Washington Post: https://github.com/washingtonpost/data-police-shootings
The Guardian: https://github.com/slanglab/policefatalities_emnlp2017/blob/master/data/gold/guardian/guard-raw.csv

## The Guardian Data Analysis and Visualization for 2016

```python
guardian = pd.read_csv("Data/guard-raw.csv")
```

```python
# code to count frequency of age ranges
one_to_twelve = 0
thirteen_to_nineteen = 0
twenties = 0
thirties = 0
forties = 0
fifties = 0
sixties = 0
seventies = 0
eighties = 0
nineties = 0

guard_age_list = []

for item in guardian["age"]:
    if item.isdigit() == True:
        # convert to integer
        item = int(item)
        guard_age_list.append(item)
        if 0 <= item <= 12:
            one_to_twelve += 1
        if 13 <= item <= 19:
            thirteen_to_nineteen +=1
        if 20 <= item <= 29:
            twenties += 1
        if 30 <= item <= 39:
            thirties += 1
        if 40 <= item <= 49:
            forties += 1
        if 50 <= item <= 59:
            fifties += 1
        if 60 <= item <= 69:
            sixties += 1
        if 70 <= item <= 79:
            seventies += 1
        if 80 <= item <= 89:
            eighties += 1
        if 90 <= item <= 99:
            nineties += 1

guard_age_dict = {"1-12": one_to_twelve, "13-19": thirteen_to_nineteen, "20-29": twenties, "30-39": thirties, \
                  "40-49": forties, "50-59": fifties, "60-69": sixties, "70-79" : seventies, "80-89": eighties, \
                  "90-99": nineties}
guard_age_dict
```

```python
df_age_guard = pd.DataFrame(guard_age_list, columns=["Age"])
df_age_guard.describe()
```

```python
keys = guard_age_dict.keys()
values = guard_age_dict.values()

plt.bar(keys, values)
plt.xlabel("Age")
plt.ylabel('Frequency')
plt.title('Frequency of Police Fatalities by Age')
plt.xticks(rotation=30)
```

## Summary of Insights into The Guardian Data

Age: The mean age in the data set is 36.656900. The median age is 35. The minimum age is 10, and the maximum age is 86. The standard deviation in ages is 12.858316. The percentage of cases in each age range are as follows: 13-19 at 5.4% (57 cases), 20-29 at 27.1% (286 cases), 30-39 at 32.9% (347 cases), 40-49 at 17% (179 cases), 50-59 at 12.1% (128 cases), 60-69 at 4.4% (46 cases), 70-79 at 0.8% (8 cases), 80-89 at 0.5% (5 cases).

Gender: 94.7% of the observtions in the data set are reported as male (1011) and 5.3% as female (57). The data set reported nobody outside of that binary, so either trans deaths by police were not reported in the news articles collected by The Guardian, or the victims were misgendered in the news articles collected.

Race/Ethnicity: Of the victims reported, 52.2% are white (558), 24.2% are Black (258), 17.1% are Hispanic (183), 2.2% are Unknown (24), 1.9% are Native American (21), 1.9% are Asian/Pacific Islander (21), 0.2% are Arab-American (3). 

Year: All observations are from the year 2016.

State: The top ten states represented in the data set are: CA (160), TX (91), FL (68), AZ (48), NC (35), CO (32), OK (31), GA (30), IL (28), OH (28).

Classification (of death): The most common classification of death in the dataset was gunshot (998) representing 93.4% of the cases. Other causes of death included 33 cases of death in custody (3%), 21 deaths by taser (1.9%), 15 victims struck by a vehicle (1.4%), and 1 case which involved some other cause of death (0.1%).

Law Enforcement Agency: The top ten agencies with the number ten spot tied are: LAPD (19), LA CO Sheriff's Dept (17), Unknown (16), Phoenix PD (15), NYPD (11), Chicago PD (11), San Antonio PD (9), US Marshals Service (8), Tulsa PD (8), CA Highway Patrol (7), and Houston PD (7).

Armed: According to the dataset, the observations reported 46.7% of victims armed with a gun (499), 14.5% armed with a knife (155), 14% not armed at all (149), 9.9% unknown whether or not the victim was armed (106), 6.6% armed with other things (70), 4.2% armed with a non-lethal firearm (45), 3.1% armed with a vehicle (34), and 0.9% disputed as to whether or not they were armed (10).

## Summary of Insights into 2016 Washington Post Data

Age: 937/962 entries in the dataset contain data on age. The mean age is 36.586980. The standard deviaton is 12.848275. The minimum age is 12. The median age is 35. The maximum age is 86. In order the percentages for each age range are as follows: 0-12 at % (1 case), 13-19 at 5.8% (54 cases), 20-29 at 27.3% (256 cases), 30-39 at 32.6% (305 cases), 40-49 at 16% (150 cases), 50-59 at 12.9% (121 cases), 60-69 at 4.1% (38 cases), 70-79 at 1% (8 cases), 80-89 at 0.4% (4 cases).

Gender: 922 victims were male (95.8%), and 40 victims were female (4.2%).

Race: The percentage of each race represented in the dataset is as follows: 51.8% White (468 cases), 25.9% Black (234 cases), 17.7% Hispanic (160 cases), 1.8% Native American (16 cases), 1.7% Asian (15 cases), 1.2% Other (11 cases).

Manner of Death: All 962 victims (93.5%) in the dataset were shot, but 62 (6.5%) were shot and tasered.

Armed: According to the dataset, the top five weapons victims were allegedly armed with were a gun (516 cases), a knife (152 cases), undetermined (57 cases), unarmed (51 cases) and a toy weapon (44 cases).

States: The top states twelve states are: CA (138), TX (82), FL (60), AZ (50), NC (33), CO (31), OK (26), OH (26), WA (26), IL (26), GA (26).

Signs of Mental Illness: Three-fourths of cases (718) showed no signs of mental illness while a quarter of cases (244) did show signs of mental illness.

Threat: 56.3% of cases allegedly had threat of attack (542), 37.7% of cases had some other threat (363), and 5.9% of cases were undetermined in having a threat (57), meaning 96% of cases allegedly posed some threat.

Flee: 66.7% Not fleeing (630), 15.1% Car (143), 12.8% Foot (121), 5.4% Other (51)

Body Camera: A body camera was only worn in 84.9% of the cases (817) and not worn in 15.1% of cases (145).

## Comparison of Data Sets for 2016

#### Entries: 
Fatal Encounters has the most entries with 1595 rows, The Guardian is second with 1068 rows, The Washington Post is less than both of these with 962 rows, and KilledByPolice.Net ties with The Washington Post with 962 entries. The difference in numbers is unclear as these datasets are all supposed to represent police fatality victims nationally.

#### Age: 
For each dataset, roughly a third of the ages fall between the 30-39 age range. Interestingly, in each data set, approximately 27% of the ages fall between the 20-29 age range. In the 40-49 age range, the percentages fall roughly +/- 1 of each other and account for approximately 16-17% of cases. The same is true for the 50-59 age range which account for about 12-13% of cases. The percentage of cases in the 60-69 age range fall between 4 and 5 percent for each dataset. The percentages for the 70-79 age range are pretty close with Fatal Encounters at 1.4%, The Guardian at 0.8%, and The Washington Post at 1%. The percentage of ages in the 80-89 age range fall between 0.4 and 0.5 percent. The mean ages for each dataset are between 36 and 37. Interestingly, The Guardian and The Washington Post both have an estimated standard deviation of 12.8 with Fatal Encounters close at 13.9. The maximum age for The Guardian and Washington Post is 86 while the maximum age for Fatal Encounters is 91. All datasets had a median age of 34-35.
Outside of the age range 13-89, the datasets begin to look vastly different. The minimum ages are all varying with Fatal ecnounters at 2, The Guardian at 10, and The Washington Post at 12. Only Fatal Encounters and The Washington Post reported police fatality victims 12 and under, but Fatal Encounters reports 15 times more cases than The Washington Post (15:1). Fatal Encounters is also the only dataset to report someone between 90-99 with one case in total.

#### Gender: 
All datasets agree that over 90% of victims in news articles are male. However, the percentage of females in The Washington Post is approximately half of that of Fatal Encounters. 
Fatal Encounters is the only dataset that accounts for trangender persons.
In the Fatal Encounters dataset, 91.8% were Male (1464), 8.1% were Female (129), and 0.8% were Transgender (1).
In The Guardian dataset, 94.7% of the observations in the data set are reported as male (1011) and 5.3% as female (57). The data set reported nobody outside of that binary, so either trans deaths by police were not reported in the news articles collected by The Guardian, or the victims were misgendered in the news articles collected.
In the Washington Post dataset, 922 victims were male (95.8%), and 40 victims were female (4.2%) in 2016.
In the KilledByPolice.Net dataset 95.8% of victims were male and 4.8% of victims were female in 2016.

#### Race:
All datasets show that approximately half of the victims represented in news articles are white. Approximately, a quarter are Black. Native Americans and Asians/Pacific Islanders represent 1.5-2% of cases, respectively. 
The percentage for Hispanic folks is approximately 17-18% in The Guardian and Washington Post data, but about 13.5% in the Fatal Encounters data. One category which cannot really be compared is the Middle Eastern category in the Fatal Encounters dataset, the Arab-American category in The Guardian dataset, and the Other category in The Washington Post dataset. It is also unclear what racial/ethnic categories fall into other. Fatal Encounters and The Guardian are the only datasets that accounts for news articles that do not specify race, but Fatal Encounters has a significantly higher number of data points which have unknown race, approximately 4:1. 

#### States:
All datasets agree that the top four states for police fatalities in order are: California, Texas, Florida, and Arizona. 
The Guardian and The Washington Post agree that North Carolina is fifth, but it is eighth in the Fatal Encounters dataset. The Guardian and The Washington Post agree that Colorado is the sixth, but Colorado is not in the top ten for Fatal Encounters at all. Oklahoma is not in the top ten for Fatal Encounters, but The Guardian and The Washington Post agree that it is in the top ten for fatalities. Georgia, Illinois, Ohio, Washington, and Oklahoma are tied in the Washington Post dataset while 
Georgia is eighth in fatalities in the The Guardian dataset with Illinois and Ohio tied. Washington state is only present in the top ten in The Washington Post dataset. Missouri and Michigan are only present in the top ten in the Fatal Encounters dataset.

Only The Washington Post and The Guardian have data on whether or not the victims were allegedly armed. Each have about 500 cases that involve victims allegedly being armed with a gun and about 150 cases of victims allegedly being armed with a knife.
